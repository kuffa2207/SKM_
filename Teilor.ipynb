  ## 1. Ряд Тейлора для синуса
Напишите функцию `mysin(x, n)`, вычисляющую сумму первых $n$
членов ряда Тейлора для функции $\sin x$:
$$
\sin x = x - \frac{x^3}{3!} + \frac{x^5}{5!} - \dots
$$
Функция должна выполнять $O(n)$ операций и не содержать питоновских циклов. 
Постройте графики получающихся функций для разных $n$.
Вариант этой функции `mysin(x)` должен вычислять столько слагаемых,
пока результат перестанет изменяться (он может содержать циклы).
Протестируйте функцию на значениях 
$x=\frac{\pi}{2},\,\frac{11\pi}{2},\,\frac{21\pi}{2},\,\frac{31\pi}{2}$.
Постройте графики зависимости точности от числа слагаемых $n$.
Какова точность результатов, полученных для таких $x$? 
Сколько членов потребовалось?
Каков максимальный по абсолютному значению член?
Попробуйте объяснить, почему результаты
неудовлетворительные.

```python
import math
import numpy as np
import matplotlib.pyplot as plt

def mysin(x, n):

    q = x # обьявляем переменные
    s = 0
    last_s = 0

    data_x = [] # создаем массивы чтобы хранить данные о сумме рядов, для отображения графика
    data_y = []
    max_q = 0

    i = 1
    flag = True
    while i < n and flag:
        q = q * (-1) * (x * x) / ((2 * i + 1) * (2 * i)) # используя формулы разложения элементарной функции sin x, выводим вот такое уравнение
        last_s = s # запоминаем последний результат суммы
        s = s + q # прибавляем q ( член ряда) к s (сумма ряда)
        if last_s == s: # проверяем совпадает ли последний результат с новым, если да то прекращаем работу
            flag = False
        if q>max_q: # проверяем больше ли наш последний член чем тот что хранится в переменной, если так то записываем его в эту переменную
            max_q = q
        data_x.append(s) # добавляем в массивы (списки) наши "точки"  то есть сумму после каждого приавления
        data_y.append(i) # добавляем в виде точек, количество n
        i += 1
    print("Потребовалось ", i, "членов")

    plt.plot(data_y, data_x) # отрисовка нашего графика
    plt.ylabel('Сумма первых n членов ряда') # текстовые метки координатных осей
    plt.xlabel('n')
    print("Максимальный по абсолютному значение член = ", max_q) # выводим в консоль
    plt.show()

mysin(math.pi/2, 100) #вызываем нашу функцию, с разными аргументами. pi/2, 11*pi/2 и т.д
mysin((11*math.pi)/2, 25)
mysin((21*math.pi)/2, 25)
mysin((31*math.pi)/2, 25)
```
